```python
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
nltk.download('stopwords')
nltk.download('vader_lexicon')
```

```python
#Define Tweet Cleaning Function
```

```python
def clean_tweet(text):
    text = str(text)
    text = text.lower()  # lowercase
    text = re.sub(r"http\S+|www\S+|https\S+", '', text, flags=re.MULTILINE)  # remove URLs
    text = re.sub(r'\@\w+|\#', '', text)  # remove mentions and hashtags
    text = re.sub(r'[^\w\s]', '', text)  # remove punctuation
    text = re.sub(r'\d+', '', text)  # remove numbers
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra spaces
    return text
```

```python
# now we can apply the function over our data
```

```python
df['cleaned_content'] = df['content'].apply(clean_tweet)
```

```python
#df['cleaned_content']
#df
```

```python
#Remove Stopwords
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    tokens = text.split()
    filtered_tokens = [w for w in tokens if w not in stop_words]
    return ' '.join(filtered_tokens)

df['no_stopwords'] = df['cleaned_content'].apply(remove_stopwords)
```

```python
#df['no_stopwords']
```

```python
#Sentiment Analysis with VADER
```

```python
sia = SentimentIntensityAnalyzer()  #Creates a VADER sentiment analyzer instance to compute sentiment scores.

def analyze_sentiment(text):       # Takes in text

    scores = sia.polarity_scores(text) #Gets sentiment scores (positive, negative, neutral, compound)

    return scores['compound'] #returns only the compound score (overall sentiment from -1 to +1)

df['sentiment_score'] = df['no_stopwords'].apply(analyze_sentiment)
```

```python
#df['sentiment_score']
```

```python
#Add a label for quick interpretation:
```

```python
def label_sentiment(score):
    if score > 0.05:
        return 'positive'
    elif score < -0.05:
        return 'negative'
    else:
        return 'neutral'

df['sentiment_label'] = df['sentiment_score'].apply(label_sentiment)
```

```python
#df['sentiment_label']
```

```python
df[['author', 'content', 'no_stopwords', 'sentiment_score', 'sentiment_label']].head()
# show the sentiment results
```

```python
import random
# Choose 3 random authors
random_authors = random.sample(df['author'].unique().tolist(), 3)

# Aggregate sentiment per selected author
sentiment_summary = df[df['author'].isin(random_authors)].groupby('author')['sentiment_score'].sum()

# Plot the sentiment per author
plt.figure(figsize=(8, 5))
sentiment_summary.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Total Sentiment Score per Author (Random Sample of 3)')
plt.xlabel('Author')
plt.ylabel('Accumulated Sentiment Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
color = ['green' if x > 0 else 'red' for x in sentiment_summary]
sentiment_summary.plot(kind='bar', color=color)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
#cleaned text
obama_text = ' '.join(df[df['author'].str.lower() == 'barackobama']['no_stopwords'].dropna())
```
